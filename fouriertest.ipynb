```python
import numpy as np

import matplotlib.pyplot as plt

# Explanation:
# The Fourier Transform decomposes a signal into its constituent frequencies.
# In numpy, np.fft.fft computes the discrete Fourier Transform (DFT) of an array.
# np.fft.fftfreq returns the frequencies corresponding to the DFT components.

# Example signal
array = np.array([1, 0, -1, 0, 1, 0, -1, 0])

# Compute the Fourier Transform
transform = np.fft.fft(array)

# Compute the corresponding frequencies
frequencies = np.fft.fftfreq(array.size, d=1)  # d=1 assumes unit sample spacing

print("Fourier Transform:", transform)
print("Frequencies:", frequencies)

# Plot the magnitude spectrum
plt.stem(frequencies, np.abs(transform.real))
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('Fourier Transform Magnitude Spectrum')
plt.show()
```

```python
# Reconstruct the original array from its Fourier transform (inverse DFT) without using a premade library


N = transform.size
reconstructed = np.zeros(N, dtype=complex)

for n in range(N):
    s = 0
    for k in range(N):
        s += transform[k] * np.exp(2j * np.pi * k * n / N)
    reconstructed[n] = s / N

# Take the real part (since the original array is real)
reconstructed_real = reconstructed.real
print(reconstructed_real)
```
